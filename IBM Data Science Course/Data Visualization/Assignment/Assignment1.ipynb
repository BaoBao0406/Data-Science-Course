```python
import numpy as np
import pandas as pd

df_t1 = pd.read_csv('Topic_Survey_Assignment.csv', index_col=0)
df_t1
```

```python
df_t1 = pd.DataFrame(df_t1.sort_values(['Very interested'], ascending=False, axis=0))

df_t1['Very interested'] = round((df_t1['Very interested'] / 2233)*100, 2)
df_t1['Somewhat interested'] = round((df_t1['Somewhat interested'] / 2233)*100, 2)
df_t1['Not interested'] = round((df_t1['Not interested'] / 2233)*100, 2)

%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_t1
```

```python
import matplotlib.patches as patches

fig, ax = plt.subplots()
interested = ['Very interested', 'Somewhat interested', 'Not interested']
df_t1[interested].plot(kind='bar', figsize=(20, 8), width=0.8, color=['#5cb85c', '#5bc0de', '#d9534f'], fontsize=14, ax=ax)

plt.title("Percentage of Respondents' Interest in Data Science Areas", fontsize=16)
plt.tick_params(labelleft='off', labelbottom='on')

for i in ax.patches:
    width, height = i.get_width(), i.get_height()
    x, y = i.get_xy()
    ax.annotate('{:.02%}'.format((height/2233)), (x, y + height + 0.01))
    

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
```
